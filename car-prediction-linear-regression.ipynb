```python
data = pd.read_csv('/kaggle/input/car-purchasing-model/Car_Purchasing_Data.csv')
data.to_parquet("data.parquet")
```

```python
from IPython.display import display

display(data)
```

```python
data.drop(["Customer Name","Customer e-mail","Country"], axis = 1, inplace = True) 
```

```python
from IPython.display import display

display(data)
```

```python
# print(data["Gender"].isnull())

for column in data:
    
    arr = data[column].isnull()
    
    for entry in arr:
        
        if entry != False:
            print("NaN")
            
print("Done")
```

Create data sets

```python
train_data = data[0:int(len(data)*.7)]
test_data = data[int(len(data)*.7):len(data)]
```

```python
from IPython.display import display

display(train_data)
```

```python
train_label = train_data["Car Purchase Amount"]
train_data = train_data[["Gender", "Age", "Annual Salary", "Credit Card Debt", "Net Worth"]]

test_label = test_data["Car Purchase Amount"]
test_data = test_data[["Gender", "Age", "Annual Salary", "Credit Card Debt", "Net Worth"]]
```

```python
from IPython.display import display

display(train_label)
display(train_data)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

lm = LinearRegression()
lm.fit(np.array(train_data), np.array(train_label))
```

```python
print(lm.score(train_data, train_label)) # Accuracy on Training Data
print(lm.score(test_data, test_label)) # Accuracy on Testing Data
```
