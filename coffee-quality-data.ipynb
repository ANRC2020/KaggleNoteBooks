```python
df = pd.read_csv("/kaggle/input/coffee-quality-data-cqi/df_arabica_clean.csv")
```

```python
df.shape
```

```python
df.drop(labels=['Unnamed: 0', 'ID', 'ICO Number', 'Status', 'Clean Cup', 'Sweetness','Defects'], inplace=True, axis=1)
```

```python
for col in df:
    if df[col].dtype == 'object':
        df[col] = df[col].factorize()[0]
```

```python
df.head()
```

```python
# Split Data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.loc[:,df.columns != "Total Cup Points"], df["Total Cup Points"], test_size=0.3)
```

```python
# Apply Linear Regression Model

from sklearn.linear_model import LinearRegression
from math import sqrt

reg = LinearRegression()

reg.fit(X_train, y_train)

preds = reg.predict(X_test)

RMSE = 0

for i, entry in enumerate(preds):
    
    print(y_test.iloc[i], entry)
    
    RMSE += abs(entry - y_test.iloc[i])**2
    
RMSE = sqrt(RMSE / y_test.shape[0])

print(f"\n{RMSE}")
```

```python
# Implement Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

model.fit(X_train, y_train)

preds = model.predict(X_test)

RMSE = 0

for i, entry in enumerate(preds):
    
#     print(y_test.iloc[i], entry)
    
    RMSE += abs(entry - y_test.iloc[i])**2
    
RMSE = sqrt(RMSE / y_test.shape[0])

print(f"\n{RMSE}")
```

```python
# Hyper Parameter Tuning for RF Model

import optuna

def objective(trial):
    
    # Split the Data Each time
    X_train, X_test, y_train, y_test = train_test_split(df.loc[:,df.columns != "Total Cup Points"], df["Total Cup Points"], test_size=0.3)
    
    # Set the Parameters
    n_estimators = trial.suggest_int('n_estimators', 2, 200)
    max_depth = int(trial.suggest_int('max_depth', 1, 32))
    
    model = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth)
    model.fit(X_train, y_train)
    
    preds = model.predict(X_test)

    RMSE = 0

    for i, entry in enumerate(preds):

    #     print(y_test.iloc[i], entry)

        RMSE += abs(entry - y_test.iloc[i])**2

    RMSE = sqrt(RMSE / y_test.shape[0])

    return RMSE
```

```python
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=200)
```

```python
trial = study.best_trial
print('RMSE: {}'.format(trial.value))
print("Best hyperparameters: {}".format(trial.params))
```
