```python
state_abbreviations = ['MP','PR','AS','VI','GU','FM','DC','MH','PW','AP','AE','AA','AL','AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN','IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH','OK','OR','PA','RI','SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY']
```

Refered to site for zip code struture info

https://guides.loc.gov/this-month-in-business-history/july/zip-code-introduced#:~:text=%E2%80%9Cthe%20five%2Ddigit%20ZIP%20number,or%20more%20less%20populated%20States.

```python
states = []
zips_first_digit = []
zips_second_digit = []
zips_third_digit = []

indices = []

for i in range(df.shape[0]):

    arr = df.iloc[i]['Address'].split(" ")
    
    for j, state in enumerate(state_abbreviations):
        
        try:
            
            index = arr.index(state)
#             print(arr[index::][0], arr[index::][1].split("-")[0])
            
            states.append(arr[index::][0])
            zips_first_digit.append(int(arr[index::][1].split("-")[0][0]))
            zips_second_digit.append(int(arr[index::][1].split("-")[0][1]))
            zips_third_digit.append(int(arr[index::][1].split("-")[0][2]))
            
            indices.append(i)
                                   
            break
        except:
            
            if j == len(state_abbreviations) - 1:
                print(arr)
            
            pass
```

```python
df['states'] = states
df['zips_first_digit'] = zips_first_digit
df['zips_second_digit'] = zips_second_digit
df['zips_third_digit'] = zips_third_digit
```

```python
df.head()
```

```python
df.drop(labels=['Address'], inplace=True, axis=1)
```

```python
df.head()
```

```python
for col in df:
    if df[col].dtype == 'object':
        df[col] = df[col].factorize()[0]
```

```python
from sklearn import preprocessing

x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.loc[:,df.columns != 5], df[5], test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
from math import sqrt

reg = LinearRegression().fit(X_train, y_train)

preds = reg.predict(X_test)

RMSE = 0

for i, entry in enumerate(preds):
    RMSE += abs(entry - y_test.iloc[i])**2
    
RMSE = sqrt(RMSE / y_test.shape[0])

print(RMSE)
```

```python
# Hyper Parameter Tuning for RF Model

import optuna

def objective(trial):
    
    # Split the Data Each time
    X_train, X_test, y_train, y_test = train_test_split(df.loc[:,df.columns != 5], df[5], test_size=0.3)
    
    # Set the Parameters
    n_estimators = trial.suggest_int('n_estimators', 2, 200)
    max_depth = int(trial.suggest_int('max_depth', 1, 32))
    
    model = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth)
    model.fit(X_train, y_train)
    
    preds = model.predict(X_test)

    RMSE = 0

    for i, entry in enumerate(preds):

    #     print(y_test.iloc[i], entry)

        RMSE += abs(entry - y_test.iloc[i])**2

    RMSE = sqrt(RMSE / y_test.shape[0])

    return RMSE
```

```python
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=200)
```

```python
trial = study.best_trial
print('RMSE: {}'.format(trial.value))
print("Best hyperparameters: {}".format(trial.params))
```
