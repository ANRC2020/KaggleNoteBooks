```python
df = pd.read_csv("/kaggle/input/water-quality-testing/Water Quality Testing.csv")
```

```python
df.isna().sum()
```

```python
df.drop(labels=['Sample ID'], inplace=True, axis=1)
```

```python
df.head()
```

```python
df.corr()
```

```python
# copy the data
df_min_max_scaled = df.copy()
  
# apply normalization techniques
for column in df_min_max_scaled.columns:
    df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())    
```

```python
df_min_max_scaled.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_min_max_scaled.loc[:,df_min_max_scaled.columns != 'pH'], df_min_max_scaled['pH'], test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
from math import sqrt

lm = LinearRegression().fit(X_train, y_train)

preds = lm.predict(X_test)

RMSE = 0

for i, pred in enumerate(preds):
    RMSE += abs(pred - y_test.iloc[i])
    
RMSE = sqrt(RMSE / len(preds))

print(RMSE)
```

```python
for i, pred in enumerate(preds):
    print(pred, y_test.iloc[i])
```

```python
from sklearn import tree

clf = tree.DecisionTreeRegressor()
clf = clf.fit(X_train, y_train)

preds = clf.predict(X_test)

RMSE = 0

for i, pred in enumerate(preds):
    RMSE += abs(pred - y_test.iloc[i])
    
RMSE = sqrt(RMSE / len(preds))

print(RMSE)
```

```python
for i, pred in enumerate(preds):
    print(pred, y_test.iloc[i])
```

```python
from sklearn.ensemble import RandomForestRegressor

clf = RandomForestRegressor()
clf.fit(X_train, y_train)

preds = clf.predict(X_test)

RMSE = 0

for i, pred in enumerate(preds):
    RMSE += abs(pred - y_test.iloc[i])
    
RMSE = sqrt(RMSE / len(preds))

print(RMSE)
```
