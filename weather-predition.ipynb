```python
df = pd.read_csv("/kaggle/input/weather-data/weather.csv")
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.drop(labels=['Unnamed: 0', 'Date'], axis=1)
```

```python
for col in df:
    if df[col].dtypes == "object":
        df[col] = df[col].factorize()[0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(df.loc[:, df.columns != "RainTomorrow"], df["RainTomorrow"], test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_train, y_train)

model.score(X_train, y_train), model.score(X_valid, y_valid)
```

```python
pred = model.predict(X_valid)
```

```python
for i, entry in enumerate(pred):
    print(entry, y_valid.iloc[i])
```
